```python
import pandas as pd

data = pd.read_csv("Perguntas e resposta - Cópia de Folha1.csv")
```

```python
data
```

```python
grouped = data.groupby("Producto")
mydict = {}
for product, group in grouped:
    print(f"Product: {product}")
    mydict[product] = {}
    perguntas = []
    resposta = []
    for _, row in group.iterrows():
        perguntas.append(row["Pergunta"])
        resposta.append(row["Resposta"])
    mydict[product]["questions"] = perguntas
    mydict[product]["ground_truths"] = resposta
```

```python
import json

# Open a file in write mode ('w')
with open("output.json", "w") as json_file:
    # Write JSON data to the file
    json.dump(mydict, json_file, indent=4)

print("JSON data has been written to output.json")
```
